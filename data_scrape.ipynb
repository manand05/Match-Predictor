```python
import soccerdata as sd

# get EPL data for a season
fbref = sd.FBref(leagues="ENG-Premier League", seasons=["2015-16","2016-17", "2017-18", "2018-19","2019-20","2020-21","2021-22","2022-23","2023-24","2024-25"])

# league match stats for all those seasons
df_matches = fbref.read_schedule()
print(df_matches.head())

# team statistics
df_team_stats = fbref.read_team_stats()
print(df_team_stats.head())
```
